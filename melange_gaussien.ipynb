First, we simulate a sample of N gaussian vectors ~ N(mean1,cov1) and a sample of N gaussian vectors ~ N(mean2,cov2).
Then we simulate a sample of latent bernoulli variable with the same size N.
This allow us to have a sample of a mixture of two gaussians.

```python
cov1=np.array([[1,0],[0,1]])
cov2=np.array([[1,0],[0,1]])
mean1=np.array((2,2))
mean2=np.array((-2,-2))
N=5000
p1=0.3
p2=0.7
d1 = np.random.multivariate_normal(mean=mean1, cov=cov1, size=N)
d2 = np.random.multivariate_normal(mean=mean2, cov=cov2, size=N)
latent = np.random.choice((True,False),N,[p1,p2])
latent_array = np.vstack((latent, latent)).T
d = np.where(latent_array, d1, d2)
```

This how our sample looks like

```python
pylab.plot(d[:,0], d[:,1], 'go')
```

Now we plot the bayes boundary decision and we plot our distribution consisting of the prepared mixture with distinctive colors.

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x, y)
Z = 0.5*np.log(np.linalg.det(cov2)/np.linalg.det(cov1))+np.log(p1/p2) + cov2[0,0]*((X-mean2[0])**2) + cov2[1,1]*((Y-mean2[1])**2) + 2*cov2[1,0]*((X-mean2[0])*(Y-mean2[1])) - (cov1[0,0]*((X-mean1[0])**2) + cov1[1,1]*((Y-mean1[1])**2) + 2*cov1[1,0]*((X-mean1[0])*(Y-mean1[1])))
```

```python
pylab.plot(d1[latent][:,0], d1[latent][:,1], 'bo', d2[np.logical_not(latent)][:,0], d2[np.logical_not(latent)][:,1], 'ro')
pylab.contour(X,Y,Z,4)
```

```python
cov1 = np.identity(2)
cov2 = np.array([[1,0.5],[0.5,1.5]])
d1 = np.random.multivariate_normal(mean=mean1, cov=cov1, size=N)
d2 = np.random.multivariate_normal(mean=mean2, cov=cov2, size=N)
latent = np.random.choice((True,False),N,[p1,p2])
latent_array = np.vstack((latent, latent)).T
d = np.where(latent_array, d1, d2)
Z = 0.5*np.log(np.linalg.det(cov2)/np.linalg.det(cov1))+np.log(p1/p2) + cov2[0,0]*((X-mean2[0])**2) + cov2[1,1]*((Y-mean2[1])**2) + 2*cov2[1,0]*((X-mean2[0])*(Y-mean2[1])) - (cov1[0,0]*((X-mean1[0])**2) + cov1[1,1]*((Y-mean1[1])**2) + 2*cov1[1,0]*((X-mean1[0])*(Y-mean1[1])))
```

```python
pylab.plot(d1[latent][:,0], d1[latent][:,1], 'bo', d2[np.logical_not(latent)][:,0], d2[np.logical_not(latent)][:,1], 'ro')
pylab.contour(X,Y,Z,4)
```

```python
def scalar_prod(u,v,matrix=None):
    if matrix is not None:
        return u.T.dot(matrix).dot(v)
    return u.T.dot(v)
```

This function defines the bayes classifier: if decision(x,y)>0 the we say that (x,y) belongs to the first population ie ~ N(mean1, cov1)

```python
def decision(x,y):
    vect = np.array((x,y))
    return (0.5*np.log(np.linalg.det(cov2)/np.linalg.det(cov1))+np.log(p1/p2)-scalar_prod(vect-mean1,vect-mean1,cov1)+scalar_prod(vect-mean2,vect-mean2,cov2))>1
```

This is the calssification accuracy:

```python
L = float(sum(decision(x,y) for x,y in d[latent]) + sum(np.logical_not(decision(x,y)) for x,y in d[np.logical_not(latent)]) )/N
L
```

## EM Algorithm:

Now we'll try to infer our density model parameters from our sample by using the EM algorithm

```python
from scipy.stats import multivariate_normal
```

```python
#Initialisation
cov1em=np.identity(2)
cov2em=np.identity(2)
mean1em=np.array((1,0))
mean2em=np.array((-1,0))
p1em=0.25
p2em=0.75
```

```python
iter_max=50
i=0
while i<iter_max:
    resp1 = np.array([p1em*multivariate_normal(mean=mean1em, cov=cov1em).pdf((x,y))\
                      /(p1em*multivariate_normal(mean=mean1em, cov=cov1em).pdf((x,y))+\
                        p2em*multivariate_normal(mean=mean2em, cov=cov2em).pdf((x,y))) for x,y in d])
    resp2 = 1 - resp1
    mean1em = np.average(d,axis=0,weights=resp1)
    mean2em = np.average(d,axis=0,weights=resp2)
    p1em = np.mean(resp1)
    p2em = 1 - p1em
    cov_term1 = np.average((d[:,0]-mean1em[0])*(d[:,1]-mean1em[1]), weights=resp1,axis=0)
    sd_term1 = np.average((d-mean1em)**2, weights=resp1,axis=0)
    cov_term2 = np.average((d[:,0]-mean2em[0])*(d[:,1]-mean2em[1]), weights=resp2,axis=0)
    sd_term2 = np.average((d-mean2em)**2, weights=resp2,axis=0)
    i=i+1
```
